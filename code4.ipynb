```python
from sklearn.cluster import MeanShift
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# بارگذاری داده‌ها
heart_disease_data = pd.read_csv("heart_disease_dataset.csv")

# انتخاب ویژگی‌های موردنظر
def selecting_data(feature1, feature2, feature3):
    features = [feature1, feature2, feature3]
    x = heart_disease_data[features].values
    return x

# اجرای الگوریتم MeanShift
def run_meanshift(band_width, x):
    # اجرای MeanShift با پهنای باند مشخص
    ms = MeanShift(bandwidth=band_width)  
    ms.fit(x)
    return ms  # برگرداندن مدل اجرا شده

# تولید برچسب‌های خوشه‌ها و مراکز
def craete_lable(ms):
    labels_meanshift = ms.labels_  # برچسب‌های خوشه‌ها
    cluster_centers = ms.cluster_centers_  # مراکز خوشه‌ها
    n_clusters = len(np.unique(labels_meanshift))  # تعداد خوشه‌ها
    print("MeanShift Cluster Labels:", labels_meanshift)
    print("Number of estimated clusters:", n_clusters)
    return labels_meanshift, cluster_centers

# رسم نمودار
def plotting(x, labels_meanshift, cluster_centers, start, end, xlabels, ylabels):
    plt.scatter(x[:, start], x[:, end], c=labels_meanshift)
    plt.scatter(cluster_centers[:, start], cluster_centers[:, end], marker='x', s=150, linewidth=5)
    plt.xlabel(xlabels)
    plt.ylabel(ylabels)
    plt.title('MeanShift Clustering')
    plt.show()


```

```python
x = selecting_data('age', 'chol', 'thalach')  
ms = run_meanshift(30, x)  
labels_meanshift, cluster_centers = craete_lable(ms)  
plotting(x, labels_meanshift, cluster_centers, 0, 1, 'Age', "Cholesterol")  
```

```python
from sklearn.cluster import MeanShift
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

heart_disease_data = pd.read_csv("heart_disease_dataset.csv")
features = ['age', 'chol', 'thalach']
x = heart_disease_data[features].values
# اجرای الگوریتم MeanShift
ms = MeanShift(bandwidth=30)  # پهنای باند: 30
ms.fit(x)

# برچسب‌های خوشه‌ها و مراکز
labels_meanshift = ms.labels_
cluster_centers = ms.cluster_centers_
n_clusters = len(np.unique(labels_meanshift))
print("MeanShift Cluster Labels:", labels_meanshift)
print("Number of estimated clusters:", n_clusters)

# رسم خوشه‌بندی بر اساس دو ویژگی
plt.scatter(x[:, 0], x[:, 1], c=labels_meanshift)
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], marker='x', s=150, linewidth=5)
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.title('MeanShift Clustering')
plt.show()
```
