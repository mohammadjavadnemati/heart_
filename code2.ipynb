```python
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
import matplotlib.pyplot as plt
import pandas as pd

# بارگذاری داده‌ها
heart_disease_data = pd.read_csv("heart_disease_dataset.csv")



def selecting_data(feature1,feature2,feature3):
    features = [feature1,feature2,feature3]
    x = heart_disease_data[features].values
    return x
def train_hir(x):

    # اجرای الگوریتم سلسله‌مراتبی
    hir = linkage(x, method='complete')  # روش لینک‌دهی: complete
    return hir
def create_lable(hir):
    # تولید برچسب‌های خوشه‌ها
    labels_hierarchical = fcluster(hir, 3, criterion='maxclust')  # تعداد خوشه‌ها: 3
    print("Hierarchical Cluster Labels:", labels_hierarchical)
    return labels_hierarchical
def plotting(hir):
    # رسم دندروگرام
    plt.figure(figsize=(10, 7))
    dendrogram(hir, leaf_rotation=90)
    plt.title("Dendrogram (Hierarchical Clustering)")
    plt.show()
def plotting2(start,end,labels_hierarchical):
    # رسم خوشه‌بندی بر اساس دو ویژگی
    plt.scatter(x[:, start], x[:, end], c=labels_hierarchical, cmap='viridis')
    plt.xlabel('Feature 1 (Standardized)')
    plt.ylabel('Feature 2 (Standardized)')
    plt.title('Hierarchical Clustering')
    plt.show()
```

```python
x=selecting_data('age', 'chol', 'thalach')
hir=train_hir(x)
labels_hierarchical=create_lable(hir)
plotting(hir)
plotting2(0,1,labels_hierarchical)
```
