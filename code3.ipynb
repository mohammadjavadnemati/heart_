```python
from sklearn.cluster import DBSCAN
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
import matplotlib.pyplot as plt
import pandas as pd


heart_disease_data = pd.read_csv("heart_disease_dataset.csv")
def selecting_data(feature1,feature2,feature3):
    features = [feature1,feature2,feature3]
    x = heart_disease_data[features].values
    return x
def train_DBSCAN(eps_arg,min_smple,x):
    # اجرای الگوریتم DBSCAN
    dbscan = DBSCAN(eps=eps_arg, min_samples=min_smple)  # شعاع: 50، حداقل نمونه‌ها: 5
    dbscan.fit(x)
    return dbscan
def create_lable(dbscan):
    # برچسب‌های خوشه‌ها
    labels_dbscan = dbscan.labels_
    print("DBSCAN Cluster Labels:", labels_dbscan)
    return labels_dbscan

def plotting(labels_dbscan,start,end,xlabels,ylabels,x):
    # رسم خوشه‌بندی بر اساس دو ویژگی
    plt.scatter(x[:, start], x[:, end], c=labels_dbscan)
    plt.xlabel(xlabels)
    plt.ylabel(ylabels)
    plt.title('DBSCAN Clustering')
    plt.show()
```

```python
x=selecting_data('age', 'chol', 'thalach')
dbscan=train_DBSCAN(50,5,x)
labels_dbscan=create_lable(dbscan)
plotting(labels_dbscan,0,1,"age","Cholesterol",x)
```

```python
from sklearn.cluster import DBSCAN
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
import matplotlib.pyplot as plt
import pandas as pd


heart_disease_data = pd.read_csv("heart_disease_dataset.csv")
features = ['age', 'chol', 'thalach']
x = heart_disease_data[features].values
# اجرای الگوریتم DBSCAN
dbscan = DBSCAN(eps=50, min_samples=5)  # شعاع: 50، حداقل نمونه‌ها: 5
dbscan.fit(x)

# برچسب‌های خوشه‌ها
labels_dbscan = dbscan.labels_
print("DBSCAN Cluster Labels:", labels_dbscan)

# رسم خوشه‌بندی بر اساس دو ویژگی
plt.scatter(x[:, 0], x[:, 1], c=labels_dbscan)
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.title('DBSCAN Clustering')
plt.show()
```
